👇 Parametered __init__ class
<pre>

```python
class car:
    def __init__(self,brand,model,price):
        self.brand = brand
        self.model = model
        self.price = price
        print("The car is from",brand,", Model is",model, ", Price is :",price)

car1 = car("BMW","M4",10000000)
car2 = car("BENZ","GWAGON",20000000)
```
