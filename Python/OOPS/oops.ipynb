```python
class car:
    def __init__(self,brand,model,price):
        self.brand = brand
        self.model = model
        self.price = price
        print("The car is from",brand,", Model is",model, ", Price is :",price)

car1 = car("BMW","M4",10000000)
car2 = car("BENZ","GWAGON",20000000)
```

👇 creating a class with multiple methods

```python
class car:
    def __init__(self,brand,model,price):
        self.brand = brand
        self.model = model
        self.price = price
        print("The car 🚘 is from",brand,", Model is",model, ", Price 💵 is :",price)

    def exterior(self,colour):
        self.colour = colour
        print("The selected car colour 🌈 option is",colour)

    def interior(self,material,colour="Not selected"):
        self.material = material
        self.colour = colour
        print("The interior 💺 material is",material, "and Interior colour is",colour)

car1 = car("Tesla","Model Y",5000000)
car1.exterior("Red")
car1.interior("Leather")
```
