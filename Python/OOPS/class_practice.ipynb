```python
class marks:
    def inp(self,mark1,mark2,mark3):
        self.mark1 = mark1
        self.mark2 = mark2
        self.mark3 = mark3

    def sum(self):
        self.total = self.mark1 + self.mark2 + self.mark3
        print(self.total)

    def avg(self):
        self.avg = self.total / 3
        print(self.avg)

m = marks()
m.inp(40,75,90)
m.sum()
m.avg()
```

```python
class Circle:
    def __init__(self):
        self.radius = int(input("Enter the radius value : "))
    def setRadius(self):
        self.value = self.radius
    def area(self):
        self.area = 3.14 * self.value ** 2
        print("Area of circle :",self.area)
    def circum(self):
        self.circum = 2 * 3.14 * self.value
        print("Circumference of the circle :",self.circum)


c = Circle()
c.setRadius()
c.area()
c.circum()
```

```python
class Student:

    def setStudDetails(self):
        self.rollNum = input("Enter the roll number of student :")
        self.studName = input("Enter the student name :")
        self.mark1 = int(input("Enter the mark1 of the student :"))
        self.mark2 = int(input("Enter the mark2 of the student :"))
        self.mark3 = int(input("Enter the mark3 of the student :"))

    def calculateTotal(self):
        print("This is the method that shows the total of student marks")
        self.totalMarks = self.mark1 + self.mark2 + self.mark3
        print("Total marks of ",self.studName,"is",self.totalMarks) 
        print("--------------------------------------------------------")

    def displayStudDetails(self):
        print("This is the method that shows the details of the student")
        print("Roll number of student :",self.rollNum)
        print("Name of the student :",self.studName)
        print("Total marks of ",self.studName,"is",self.totalMarks) 
        print("--------------------------------------------------------")

class StudentDemo:
    def __init__(self):
        s = Student()
        s.setStudDetails()
        s.calculateTotal()
        s.displayStudDetails()

demo = StudentDemo()
```

```python
class Account:
    def __init__(self):
        self.accountNo = 0
        self.accountType = ""
        self.accountBalance = 0

    def setAccountDetails(self,acno,acty,acbl):
        self.accountNo = acno
        self.accountType = acty
        self.accountBalance = acbl

    def withdraw(self,withdrawalAmount):
        self.finalBalance = self.accountBalance - withdrawalAmount
        print("After withdrawal of",withdrawalAmount,"your account balance is :",self.finalBalance)
        print("--------------------------------------------------------")
        

    def deposit(self,depositAmount):
        self.finalBalance = self.accountBalance + depositAmount
        print("After depositing",depositAmount,"your account balance is :",self.finalBalance)
        print("--------------------------------------------------------")

    def dispAccountDetails(self):
        print("Your account number is :",self.accountNo)
        print("Your account type is :",self.accountType)
        print("Your account balance is :",self.accountBalance,"Before any transaction, After transaction",self.finalBalance) 
        print("--------------------------------------------------------")

class Main1:
   def main(self):
    Ac = Account()
    Ac.setAccountDetails(987654321,"Savings",703300) 
    Ac.withdraw(2000)
    Ac.deposit(1345)
    Ac.dispAccountDetails()

m = Main1()
m.main()
```
