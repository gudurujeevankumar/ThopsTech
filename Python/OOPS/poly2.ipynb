🟢 1. Method Overloading (Compile-time concept in theory)

❓ Problem:

Create a class Calculator with method add() such that:
	•	If 2 arguments → return sum of 2 numbers
	•	If 3 arguments → return sum of 3 numbers

👉 (Note: Python doesn’t support true overloading, so simulate it)

```python
from typing_extensions import Self
class Calculator:
    def add(self,a,b,c = None):
        if c == None:
            return a + b
        return a + b + c

c = Calculator()
print("If we give 2 numbers to add method : ",c.add(3,5))
print("If we give 3 numbers to add method :",c.add(5,6,3))
```

🔵 2. Runtime Polymorphism (Method Overriding)

❓ Problem:

Create a base class Employee with method salary().

Create subclasses:
	•	Developer → salary = base + bonus
	•	Manager → salary = base + incentives

👉 Use same method salary() but different implementation

```python
class Employee:
    def salary(self):
        pass

class Developer(Employee):
    def __init__(self,base,bonus):
        self.base = base
        self.bonus = bonus

    def salary(self):
        totalSalary = self.base + self.bonus
        print("salary method in Developer sub class")
        return totalSalary

class Manager(Employee):
    def __init__(self,base,insentives):
        self.base = base
        self.insentives = insentives

    def salary(self):
        totalSalary = self.base + self.insentives
        print("salary method in Manager sub class")
        return totalSalary
        

e = Employee()
d = Developer(1240300,30200)
m = Manager(1523500,50300)
print("Total Salary of manager : ",m.salary())
print("----------------------------------------")
print("Total Salary of developer : ",d.salary())
```
