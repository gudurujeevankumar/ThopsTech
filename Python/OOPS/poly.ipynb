1. Method Overriding

Create:
	•	Animal → sound()
	•	Dog, Cat override it

```python
class Animal:
    def sound(self):
        print("This is base class having sound method")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("cat Meow")


animals = [Animal(), Dog(), Cat()]

for animal in animals:
    animal.sound()
```

2. Shape Area

Create:
	•	Shape → area()
	•	Circle, Rectangle → implement differently

```python
class Shape:
    def area(self):
        print("This is base class method that calculate the area")

class Circle(Shape):
    def area(self,r):
        self.circleArea = 3.14 * r ** 2
        print("Area of the circle is : ", self.circleArea)

class Rectangle(Shape):
    def area(self,l,w):
        self.recArea = l * w
        print("Area of rectangle is : ",self.recArea)

s = Shape()
c = Circle()
r = Rectangle()

s.area()
c.area(5.3)
r.area(4,2.2)
```

3. Payment System

Base: Payment → pay()

Subclasses:
	•	CreditCard
	•	UPI
	•	Cash

```python
class Payment:
    def pay(self):
        pass

class CreditCard(Payment):
    def pay(self):
        print("Paying through credit card")

class UPI(Payment):
    def pay(self):
        print("Paying through UPI")

class Cash(Payment):
    def pay(self):
        print("Paying through cash")

payment_ways = [Payment(),CreditCard(),UPI(),Cash()]

for ways in payment_ways:
    ways.pay()
```
