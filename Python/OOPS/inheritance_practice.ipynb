```python
class Person:
    def __init__(self):
        self.name = ""
        self.age = 0

    def input(self):
        self.name = input("Enter Name: ")
        self.age = int(input("Enter Age: "))

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)


class Student(Person):
    def __init__(self):
        super().__init__()
        self.marks = 0

    def input(self):
        super().input()
        self.marks = int(input("Enter Marks: "))

    def display(self):
        super().display()
        print("Marks:", self.marks)


def main():
    s = Student()
    s.input()
    s.display()


main()
```

```python
class Employee:
    def __init__(self):
        self.empId = ""
        self.name = ""
    def inpEmp(self):
        self.empId = input("Enter the employee ID : ")
        self.name = input("Enter the employee name : ")

class Salary(Employee):
    def __init__(self):
        super().__init__()
        self.basic = 0
    def inpSal(self):
        self.inpEmp()
        self.basic = int(input("Enter the basic sallary of employee : "))
        
class Bonus(Salary):
    def __init__(self):
        super().__init__()
        self.bonus = 0
    def inpBon(self):
        self.inpSal()
        self.bonus = int(input("Enter the bonus : "))
    def totalSalary(self):
        self.total = self.basic + self.bonus
        print("Total sallary after adding bonus : ",self.total)

    def display(self):
        print("------------------------------")
        print("Employee name : ",self.name)
        print("Employee ID : ",self.empId)
        print("Employee Basic Sallary : ",self.basic)
        print("Employee Bonus Sallary : ",self.bonus)
        print("Employee Total Sallary : ",self.total)

def main():
    b = Bonus()
    b.inpBon()
    b.totalSalary()
    b.display()

main()

```
